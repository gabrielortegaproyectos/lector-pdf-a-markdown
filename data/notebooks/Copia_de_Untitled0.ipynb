# Conversión de PDF a Markdown con Marker-PDF

Este cuaderno de Colab te guiará paso a paso para convertir un archivo PDF en un formato Markdown usando la biblioteca `marker-pdf`. `marker-pdf` es una herramienta potente que utiliza modelos de aprendizaje profundo para extraer el contenido y la estructura de documentos PDF, transformándolos en texto Markdown legible.

Sigue los siguientes pasos para cargar tu PDF y obtener su versión en Markdown.

## Paso 2: Subir tu archivo PDF

Utiliza el siguiente explorador de archivos para subir el PDF que deseas convertir. Una vez que hagas clic en 'Ejecutar' en la celda de código de abajo, aparecerá un botón 'Examinar' (o 'Choose Files'). Haz clic en él, navega hasta donde esté tu archivo PDF en tu computador y selecciónalo. Una vez subido, el nombre del archivo se almacenará en la variable `pdf_filename`.

**Importante:** Asegúrate de subir un archivo con extensión `.pdf`.

```python
from google.colab import files
import os

uploaded = files.upload()

pdf_filename = None
if uploaded:
    # Get the name of the first (and likely only) uploaded file
    pdf_filename = list(uploaded.keys())[0]
    print(f"Archivo '{pdf_filename}' subido exitosamente.")
else:
    print("No se ha subido ningún archivo.")

# Verificar si el archivo es un PDF
if pdf_filename and not pdf_filename.lower().endswith(".pdf"):
    print("Advertencia: El archivo subido no parece ser un PDF. Asegúrate de subir un archivo .pdf.")
    # You might want to clear pdf_filename if it's not a PDF to prevent errors later
    pdf_filename = None
```

## Paso 3: Cargar los modelos de Marker-PDF y convertir el PDF

Aquí importamos las clases necesarias de `marker-pdf`, cargamos los modelos de inteligencia artificial preentrenados y realizamos la conversión del PDF. La primera vez que se cargan los modelos (`create_model_dict()`), puede tardar un poco ya que se descargan los recursos necesarios.

El `PdfConverter` toma el archivo PDF (el que subiste en el paso anterior) y lo procesa para extraer su estructura y contenido. Luego, `text_from_rendered` convierte esta representación interna en una cadena de texto en formato Markdown.

```python
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered

# Cargar modelos (solo la primera vez tarda)
# Este paso descarga los modelos de IA necesarios para la conversión.
# Puede tomar unos minutos la primera vez que se ejecuta.
models = create_model_dict()

# Convertir
converter = PdfConverter(artifact_dict=models)

markdown = "" # Initialize markdown variable

if 'pdf_filename' in locals() and pdf_filename:
    print(f"Convirtiendo el archivo '{pdf_filename}' a Markdown...")
    rendered = converter(pdf_filename)

    # Obtener el markdown
    markdown, _, _ = text_from_rendered(rendered)
    print("\n--- Contenido Markdown generado (fragmento) ---")
    print(markdown[:1000]) # Print only first 1000 characters for brevity
    print("------------------------------------------------\n")
elif 'pdf_filename' not in locals():
    print("La variable 'pdf_filename' no está definida. Por favor, asegúrate de subir un archivo PDF en el Paso 2.")
else:
    print("No se puede realizar la conversión porque no se ha subido ningún archivo PDF válido en el Paso 2.")
```

## Paso 4: Guardar el Markdown resultante en un archivo

Finalmente, guardamos el texto Markdown generado en un archivo local llamado `output.md`. Puedes descargar este archivo desde el panel de archivos de Colab (el icono de la carpeta en la barra lateral izquierda) una vez que la celda se haya ejecutado.

```python
if 'markdown' in locals() and markdown:
    with open("output.md", "w", encoding="utf-8") as f:
        f.write(markdown)

    print("¡Listo! El contenido Markdown se ha guardado en 'output.md'.")
else:
    print("No se generó contenido Markdown para guardar. Asegúrate de que la conversión se haya realizado correctamente en el Paso 3.")
```

## Conclusión

¡Felicidades! Has convertido exitosamente tu documento PDF a Markdown. Ahora puedes usar el archivo `output.md` para editar, publicar o integrar el contenido en otras plataformas.

```python
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered

# Cargar modelos (solo la primera vez tarda)
models = create_model_dict()

# Convertir
converter = PdfConverter(artifact_dict=models)
rendered = converter("Doble.pdf")

# Obtener el markdown
markdown, _, _ = text_from_rendered(rendered)
print(markdown)
```

```python
with open("output.md", "w", encoding="utf-8") as f:
    f.write(markdown)

print("¡Listo! Guardado en output.md")
```
